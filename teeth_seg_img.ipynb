```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
images_path = 'drive/MyDrive/Colab Notebooks/cat6/cropped_images/'
train_images_path = 'drive/MyDrive/Colab Notebooks/cat6/cropped_images/annotated_images/'

training_images = os.listdir(train_images_path)
org_images = os.listdir(images_path)
training_images.remove('mouth')
org_images.remove('segmented_images')
org_images.remove('annotated_images')
```

```python
img = cv2.imread(images_path + org_images[0])
img_gry = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.subplot(1,1,1)
plt.imshow(img_rgb)
plt.show()
import shutil
#shutil.rmtree('../results')
```

```python
img1 = cv2.imread(images_path + org_images[0])
mask = cv2.imread(train_images_path+training_images[0])
masked_image = cv2.bitwise_and(img1,mask)

if os.path.exists('/content/drive/MyDrive/results') ==False:
  os.mkdir('/content/drive/MyDrive/results')
#cv2.imshow(masked_image)
plt.imshow(masked_image)
cv2.waitKey(0)
for i in range(len(org_images)):
  img1 = cv2.imread(images_path + org_images[i])
  mask = cv2.imread(train_images_path+training_images[i])
  result = cv2.bitwise_and(img1,mask)
  plt.savefig(f'/content/drive/MyDrive/results/result{i}')
```
